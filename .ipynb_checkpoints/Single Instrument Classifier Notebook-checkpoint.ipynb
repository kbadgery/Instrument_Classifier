# Strategy
When a note is played on an instrument, a sound wave is generated. This wave contains both the frequency of the note, as well as a number of subfrequencies that are unique to that instrument. The combination of these sub-frequencies, commonly referred to as "timbre," is what gives each instrument it's unique character. 

What makes this problem unique, is that our ability to predict what instrument is being played is not determined by the specific frequencies in that sound file, but rather how those frequncies are spaced relative to one another. For example it is of no use for us to know that the frequency response has as spike at 780hz. However, knowing that there are spikes at 780hz, 1100hz, and 1280hz, is useful. In other words, it is not the frequencies themselves that are useful for classification, but rather the patterns in which these frequencies appear (more specifically, the spacing of these frequencies on a logarithmic scale). 

In order to capture these spacing patterns, I will use a convolutional neural network (CNN). CNN's are useful for extracting spatial data (in this case the "space" is the frequency response of a sound wave). At a basic level, CNN's apply filters to data in order to extract certain information. The output of this filtered data is then fed to a standard fully-connected neural network. CNN's are most commonly known for image recognition, another classification task that is dependant on the spatial relationship between the pixels in an image, not the values of the pixels by themselves. In other words, the value of a single pixel won't give you any information on what is in an image, but the relationship between a number of pixels will.

# Rebucketing the DFT

Next I rebucket the fft by creating 1000 logarithmically spaced buckets and assigning the amplitude of each data point from the original fft to it's nearest bucket, I then sum these amplitudes to create a new rebucketed frequency response as seen below. *Note: This isn't technically how a discrete fourier transform should be rebucketed, but for our purposes it is sufficient*

```python
freq_bins = np.geomspace(1, 15000, num=1000) # creates 1000 logarithmically spaced frequency buckets
rebucketed_fft = reBucket(fft_amps, freq_vector, freq_bins)

plt.figure()
plt.semilogx(freq_bins, rebucketed_fft)
```

# Network Architecture

## 1-D Convolution layer

### The Convolution Process

As mentioned earlier, I will use a CNN to learn from thousands of instances, similar to the one above. To capture useful information hiding in this data, the CNN iteratively passes filters over each instance.

We begin by starting the filter on the far left side of the modified frequency response and calculating the dot product. We then iterate by shifting the filter to the right. Normally, this is done until the filter can no longer shift to the right, but for simplicity I have only shown the first 3 iterations iteration.

By calculating the dot product for each iteration, we can get an idea of the "similarity" between the filter and the frequency response (for that iteration).

```python
eg_data = [0,0,2,400,8,130,5,0,0]
eg_filters = [[5,300,5,100,5,0,0,0,0],[0,5,300,5,100,5,0,0,0],[0,0,5,300,5,100,5,0,0]]
           
for filt in eg_filters:    

    plt.figure()
    plt.plot(eg_data)
    plt.plot(filt)
    plt.title("Iteration #" + str(eg_filters.index(filt)+1) + " - Dot Product = " + str(sum([x*y for x,y in zip(eg_data,filt)])))
    plt.legend(['Data','Filter'])
```

Here we can see that the peaks line up on the 3rd iteration, while on the 1st and 2nd iteration this is not the case. As a result, we get a much higher dot product. This is why the dot product is useful for determining the "similarity" between the two vectors.

Now to apply this to the problem of instrument classification. To classify our instruments we could run a series of filters over the fourier transforms of instruments, and the high dot products will indicate similarity between filter and data. By picking filters that are similar to the fourier transforms of the instruments we are interested in, we can begin to classify instruments. A filter that is similar to the fourier transform of a guitar note would likely result in a high dot product with a guitar than it would with a keyboard. 

However this raises the problem: how do we pick these filters, and how do we figure out which instruments they correspond to. 

Luckily, this is what CNNs are great at! Just like how other machine learning algorithms are able to learn the relationship between input features and output results, the CNN uses backpropogation to learn which filters are most useful, and also learn how to use the output of the convolution operation to predict each instrument. 

*Note: Unlike the filter shown in the figures above, the actual filters that the CNN learns may not be immediately intuitive to a human, instead, the algorithm will simply learn whichever filters allow it to make the best prediction. Then, by combining the outputs from the many seperate filter convolutions, the algorithm is able to learn useful features of each individual instrument.*

### Filter Size

I will be using a filter size that is equal to the number of buckets of the fourier transform (1000). The intuition behind this is that the useful sub-frequncies combinations that are unique to each instrument could exist anywhere within the fourier transform. If two frequncies that are useful for prediction exist at opposite ends of the fourier transform, then we need a filter that spans the entire fourier transform to capture this information.

*Note: Ideally this would be a hyperparameter that would be optimized in the bayesian optimization stage. Due to computational constraints, this was not tested.*

## Global max pooling layer

I use a global max pooling layer to reduce the dimensionality of the data that leaves the convolution layer. In the example figures above, we can see that we would not need the output from each dot product to be able to predict the instrument. Instead we would just need to know the output of the iteration with the highest dot product.

To achieve this, I use a global max pooling layer to select only the highest result for each filter, and pass this value on to the fully connected layers. As a result, the length of the output vector of this layer is equal to the number of filters.

## Fully connected layers

The output of the global max pooling layer is then fed into a standard neural network architecture comprised of 2 fully connected layers of neurons. 

This standard neural network architecture is used in many machine learning tasks. I will not go into much detail here as I am using a very standard neural network architecture for this part of the algorithm. 

In general, neural networks are good at learning complex relationships between input features (a measureable property that hopefully has predictive power) and a model output. For example, a neural network designed to predict house prices may use features such as the square footage or number of bedrooms to predict the house price. One of the main things that distinguishes neural networks from other machine leanring algorithms is the neural networks superior ability to learn complex absract relationships between features, and use this to improve predictive power (e.g. perhaps the square footage has less of an impact on house price if the number of bedrooms is low).


## Hyperparameter Optimization
Neural networks have two types of parameters that can be altered:

Parameters: The weights and biases of the nodes in the network, these are initialized randomly, and are then tuned by training the neural network using backpropogation.

Hyperparameters: The network parameters that must be set prior to training, and are not optimized by the training of the network. The list of types of hyperparmeters is long, and includes things such as: the number of neurons in each layer, the number of layers, the types of activation functions used in the layers, the batch size, and many, many more. These hyperparameters are what create the network architecture.


Ideally, we want to select the set of hyperparameters that results in the best performance possible. The problem is, the number of possible sets of hyperparameters (aka the hyperparameter space) is generally very large for neural networks. On top of this, because neural networks are "black boxes," the only tool we have to test any set of hyperparameters, is to create a model with those hyperparameters, train it, and evaluate it. This is incredibly computationally expensive, and generally it is not computationally feasible to sample the entire hyperparameter space, particularily when working with large networks or big datasets.

In this project, the final set of hyperparameters was found using bayesian hyperparameter optimization (I did this seperately because it takes a really long time to run, the spyder notebook for which can be found in this github folder). 

Bayesian hyperparameter optimization is a specific use case of bayesian optimization that can be applied to the optimization of hyperparameters in a machine learning algorithm.

Unlike other more basic hyperparameter optimization techniques such as grid search and random grid search, bayesian hyperparameter optimization doesn't attempt to search the entire hyperparameter space, or randomly sample from it. Instead, it calculates a gaussian distribution of possible performance outcomes for each point in the hyperparameter space. It then uses these distributions to select the set of hyperparameters that creates the highest level of expected improvement. This set is then tested, and used to update the gaussian distributions of all other points in the hyperparameter space.

For this, I used this great library: https://github.com/fmfn/BayesianOptimization, which can be used to apply bayesian optmization to any function.

*Note: There are a number of hyperparamters I chose not to try to optimize. Examples include the number of buckets used in the input fourier transform, and the size of the filter. This was done due to computational constraints and a more comprehensive analysis would include this at this stage.*

# Importing Libraries and Reading the data

```python
from keras.models import Sequential, clone_model
from keras.layers import Conv1D, GlobalMaxPooling1D, MaxPooling1D, Flatten, Dense, Dropout
from keras.optimizers import Adam
from keras import regularizers
from keras.metrics import Recall
from sklearn.model_selection import train_test_split
from sklearn.metrics import recall_score, confusion_matrix
```

```python
X = np.load("Processed Data/single_instruments_1000_X.npy")
y = np.load("Processed Data/single_instruments_1000_y.npy")

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size = 0.2, random_state = 2)
```

# Building the Model

Here we use the set of hyperparameters found using bayesion opitimization to construct our model in Keras

Best set:
batch size: 549
no regularization
learning rate: 0.00086
n_epochs: 20
n_filters: 114
n_layers: 2
neurons/layer: 127


```python
# Model Architecture
classifier = Sequential()
classifier.add(Conv1D(114, 1000, input_shape = (len(X[0]), 1), strides=1, padding='same',activation = 'relu'))
classifier.add(GlobalMaxPooling1D())
classifier.add(Dense(units = 127, activation = 'relu'))
classifier.add(Dense(units = 127, activation = 'relu'))
classifier.add(Dense(units = len(y[0]), activation = 'softmax'))

# Compiling
opt = Adam(learning_rate=0.00086)
classifier.compile(optimizer = opt, loss = 'categorical_crossentropy', metrics = ['accuracy','mean_squared_error'])
```

# Training

Now we use the data that we preprocessed to train the model

```python
recall_train = []
recall_val = []   
```

```python
n_epochs = 250
best_recall = 0

for i in range(n_epochs):
    classifier.fit(X_train, y_train, batch_size = 549, epochs = 1, verbose=0)
    
    y_pred_train = classifier.predict(X_train)
    recall_train.append(IC_recall(y_pred_train, y_train))
    
    y_pred_val = classifier.predict(X_val)
    current_recall_val = IC_recall(y_pred_val, y_val)
    recall_val.append(current_recall_val)
    
    if current_recall_val > best_recall:
        best_model = clone_model(classifier)
        best_model.set_weights(classifier.get_weights())
        best_recall = current_recall_val
```

```python
plt.figure()
plt.plot(recall_train)
plt.plot(recall_val)
plt.plot(recall_val.index(max(recall_val)),max(recall_val),'o')
plt.ylim(0.8, 1)
plt.title("Model Accuracy")
plt.xlabel("Number of Epochs")
plt.ylabel("Accuracy")
plt.legend(['Training Set','Validation Set','Best Result'])
```

As we can see, the model may not have fully converged, and instead was limited by the fact that it was slightly overfitting the training data (We know this because the training accuracy was higher than the validation accuracy). Further work could look at reducing overfitting by using more training data, or a regularization technique such as L1, L2, or dropout layers.
